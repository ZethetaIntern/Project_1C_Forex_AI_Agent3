```python
import pandas as pd

mc = pd.read_csv("../data/mc_scenarios.csv")

mc.info()

print(mc.columns.tolist())
```

```python
# Best, Worst and Average Portfolio Performance

best_case = mc["PortfolioPnL"].max()

worst_case = mc["PortfolioPnL"].min()

expected_return = mc["PortfolioPnL"].mean()

print(f"Best Case Portfolio PnL : {best_case:.2f}")

print(f"Worst Case Portfolio PnL : {worst_case:.2f}")

print(f"Expected Portfolio PnL : {expected_return:.2f}")
```

```python
plt.figure(figsize=(12,6))

plt.hist(
    mc["PortfolioPnL"],
    bins=30
)

plt.title("Monte Carlo Portfolio PnL Distribution")

plt.xlabel("Portfolio Profit / Loss")

plt.ylabel("Number of Scenarios")

plt.grid(True)

plt.show()
```

```python
# Value at Risk (95%)

var_95 = np.percentile(
    mc["PortfolioPnL"],
    5
)

print(f"95% Value at Risk (VaR): {var_95:.2f}")
```

```python
# Value at Risk (95%)

var_95 = np.percentile(
    mc["PortfolioPnL"],
    5
)

print(f"95% Value at Risk (VaR): {var_95:.2f}")
```

```python
# Expected Shortfall (Conditional VaR)

cvar_95 = mc.loc[
    mc["PortfolioPnL"] <= var_95,
    "PortfolioPnL"
].mean()

print(f"95% Expected Shortfall (CVaR): {cvar_95:.2f}")
```
