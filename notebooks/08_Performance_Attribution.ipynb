```python
import numpy as np

# Win Rate
win_rate = (trade["PnL"] > 0).mean() * 100

# Profit Factor
gross_profit = trade.loc[trade["PnL"] > 0, "PnL"].sum()
gross_loss = abs(trade.loc[trade["PnL"] < 0, "PnL"].sum())

profit_factor = gross_profit / gross_loss if gross_loss != 0 else np.nan

# Sharpe Ratio
sharpe_ratio = trade["PnL"].mean() / trade["PnL"].std()

summary = pd.DataFrame({
    "Metric":[
        "Win Rate (%)",
        "Profit Factor",
        "Sharpe Ratio",
        "Total Trades"
    ],
    "Value":[
        round(win_rate,2),
        round(profit_factor,2),
        round(sharpe_ratio,2),
        len(trade)
    ]
})

summary
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
trade = pd.read_csv("../data/trade_log.csv")

trade.head()
```

```python
pair_performance = trade.groupby("PairName")["PnL"].sum()

print(pair_performance)
```

```python
plt.figure(figsize=(8,5))

pair_performance.plot(kind="bar")

plt.title("Profit by Currency Pair")

plt.xlabel("Currency Pair")

plt.ylabel("Total PnL")

plt.grid(True)

plt.show()
```
