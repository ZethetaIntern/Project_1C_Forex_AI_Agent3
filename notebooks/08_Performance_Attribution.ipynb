```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
trade = pd.read_csv("../data/trade_log.csv")

trade.head()
```

```python
pair_performance = trade.groupby("PairName")["PnL"].sum()

print(pair_performance)
```

```python
plt.figure(figsize=(8,5))

pair_performance.plot(kind="bar")

plt.title("Profit by Currency Pair")

plt.xlabel("Currency Pair")

plt.ylabel("Total PnL")

plt.grid(True)

plt.show()
```

```python
strategy_performance = trade.groupby("StrategyName")["PnL"].sum()

print(strategy_performance)
```

```python
strategy_performance = trade.groupby("StrategyName")["PnL"].sum()

print(strategy_performance)
```

```python
direction_performance = trade.groupby("Direction")["PnL"].sum()

print(direction_performance)
```

```python
plt.figure(figsize=(6,5))

direction_performance.plot(kind="bar")

plt.title("BUY vs SELL Performance")

plt.ylabel("Total PnL")

plt.grid(True)

plt.show()
```

```python
total_commission = trade["Commission"].sum()

average_commission = trade["Commission"].mean()

print("Total Commission:", total_commission)

print("Average Commission:", average_commission)
```

```python
average_holding = trade["HoldingDays"].mean()

print("Average Holding Period:", average_holding)
```
